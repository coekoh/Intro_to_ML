```python
#LogisticRegression():
def sigmoid(z):
    result = 1/(1 + np.exp(-z))
    result[result< 1e-16] = 1e-16
    return result

def cross_entropy(X, y, w, b, lamda):
    p = sigmoid(X @ w + b)
    entropy = y*np.log(p) + (1-y)*np.log(1 - p)
    total = entropy.sum()
    return -total + (lamda*np.square(np.linalg.norm(w)))
    
def initilize_parameters(n_x):
    w = np.zeros((n_x, 1))
    b = 0.1
    return w, b
    
def fit(X, w, b):
    A = X@w + b
    return sigmoid(A)
    
def gradient(X, y, w, b, lamda):
    dw =(np.dot(X.T,sigmoid(X@w + b) - y)/len(X)) + lamda*w
    db = np.sum(sigmoid(X@w + b) - y)/len(X)
    return dw, db
        
def update(dw, db, w, b, learning_rate):
    w = w - learning_rate*dw
    b = b - learning_rate*db
    return w, b
```

```python
X_train = df.drop(columns = ["y"]).to_numpy()
y_train = df["y"].to_numpy().reshape(-1, 1)
```

### Question 4.4b

```python
learning_rates = [.001, .01, .05, 0.1, 0.5]
```

```python
fig, ax = plt.subplots(figsize = (12, 8))

norm = []
for i in learning_rates:
    cross = []
    w, b = initilize_parameters(X_train.shape[1])
    for j in range(50):
        dw, db = gradient(X_train, y_train, w, b, lamda = 0)
        w, b = update(dw, db, w, b, i)
        cross.append(cross_entropy(X_train, y_train, w, b, lamda = 0))
    norm.append(np.linalg.norm(w))
    ax.plot([i for i in range(1, 51)], cross, label = "learning rate: " + str(i))
    
plt.ylabel("Losses")
plt.xlabel("Number of Iterations")
plt.title("Plot of the cross-entropy function value with respect to" + 
          "the number of iterations (T = [1,..,50]) for the training data for each step size.")
plt.legend()
print("The L2 norms after every Learning rate are: ")
print(norm)
plt.show()
```
